# Do priors matter??

Inspired by https://twitter.com/rlmcelreath/status/1701165078949581182/photo/1

Use `scipy.stats` to generate four distributions:
- a standard normal with mean 0, variance 1
- a standard normal with mean 10, variance 1
- a student's with 2 degrees of freedom centered on 0
- a student's with 2 degrees of freedom centered on 10

Plot all the distributions


Create a function that takes a prior, a likelihood as arguments and plots the prior, likelihood, and posterior (normalize each so that the peak is at the same height for optimal visibility)

Pass to the function each combination of prior and likelihood between the four distributions you creted: normal-normal, t-t, normal-t, t-normal



```python
#these are all the imports you need
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(18)
```

```python
x = np.linspace(-5,15, 1000)
```

```python
#from class
N_0=stats.norm.pdf(x)
N_10=stats.norm.pdf(x,loc=10)
t_0=stats.t.pdf(x, 2,loc=0,scale=1)
t_10=stats.t.pdf(x, 2,loc=10,scale=1)
plt.plot(x,N_0,label="N(0,1)")
plt.plot(x,N_10,label="N(10,1)")
plt.plot(x,t_0,label="t(0,1)")
plt.plot(x,t_10,label="t(10,1)")
plt.legend()
```

```python
fig, ax = plt.subplots(1, 1)


ax.plot(x, stats.norm.pdf(x))
plt.title("standard normal with mean 0, variance 1")
fig.show()
```

```python
plt.plot(x, stats.norm.pdf(x, loc=10, scale=1))
plt.xlim(0,20)
plt.title("standard normal with mean 10, variance 1")
```

```python
#plt.plot(x, stats.norm.pdf(x, loc=10, scale=1))
plt.xlim(0,20)
plt.title("student's with 2 degrees of freedom centered on 0")
```

```python
"""
Create a function that takes a prior, a likelihood as arguments and
plots the prior, likelihood, and posterior
(normalize each so that the peak is at the same height for optimal visibility)
"""
def posterior(p, l):
  plt.plot(x, p/p.max(), "k--", label="prior")
  plt.plot(x, l/l.max(), "k-", label="likelihood")
  post = p*l
  plt.plot(x, post/post.max(), "red", label="posterior")
  plt.legend()
```

```python
plt.title("prior of normal at 0, likelihood of normal at 10")
posterior(N_0,N_10)
```

```python
"""Pass to the function each combination of prior and likelihood
between the four distributions you creted:
normal-normal, t-t, normal-t, t-normal
"""
```

```python
plt.title("prior of normal at 10, likelihood of normal at 0")
posterior(N_10,N_0)
```

```python
plt.title("prior of t at 10, likelihood of t at 0")
posterior(t_10,t_0)
```

```python
plt.title("prior of t at 0, likelihood of t at 10")
posterior(t_0,t_10)
```

```python
plt.title("prior of normal at 10, likelihood of t at 0")
posterior(N_10,t_0)
```

```python
plt.title("prior of normal at 0, likelihood of t at 10")
posterior(N_0,t_10)
```
